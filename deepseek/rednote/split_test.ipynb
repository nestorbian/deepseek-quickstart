```python
import re

# 读取文件内容
with open('realdata.md', 'r', encoding='utf-8') as file:
    content = file.read()

# 使用正则表达式分割产品
# 模式解释：匹配以 "# 数字. " 开头的行作为产品分隔符
# re.MULTILINE 确保 ^ 匹配每行开头，re.DOTALL 使 . 匹配换行符
pattern = r'(^# \d+\. .+?$)'  # 匹配产品标题行
products = re.split(pattern, content, flags=re.MULTILINE | re.DOTALL)
```

```python
# 过滤空字符串并配对标题和内容
# 分割后奇数索引是标题，偶数索引是内容
product_list = []
for i in range(1, len(products)):
    if i % 2 == 1:  # 标题索引（1,3,5...）
        title = products[i].strip()
        content = products[i+1].strip() if i+1 < len(products) else ""
        product_list.append((title, content))

# 打印结果
for idx, (title, content) in enumerate(product_list, 1):
    print(f"产品 {idx}:")
    print("标题:", title)
    print("内容:", content[:50] + "..." if len(content) > 50 else content)  # 预览部分内容
    print("-" * 50)

# # 如果需要保存到单独文件
# for idx, (title, content) in enumerate(product_list, 1):
#     filename = f"product_{idx}.md"
#     with open(filename, 'w', encoding='utf-8') as out_file:
#         out_file.write(title + "\n" + content)
#     print(f"已保存: {filename}")
```
